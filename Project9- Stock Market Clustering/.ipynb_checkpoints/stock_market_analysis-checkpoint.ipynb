#### **Normalization**
Stock prices of Amazon and Apple have diffirent scales. There is a significant diffirence in the rate of change of units in stock price of Amazon and Apple and in that case we need to normalise our data in order to have consistency in our data for further analysis.


```python
plt.figure(figsize=(20,8))

plt.subplot(1,2,1)

plt.title('Company: Amazon', fontsize= 20)
plt.xticks(fontsize=10)
plt.yticks(fontsize=20)
plt.xlabel('Date', fontsize=20)
plt.ylabel('Movement',fontsize=20)
plt.plot(movements[0]) #0 for Amazon

plt.subplot(1,2,2)

plt.title('Company: Apple', fontsize= 20)
plt.xticks(fontsize=10)
plt.yticks(fontsize=20)
plt.xlabel('Date', fontsize=20)
plt.ylabel('Movement',fontsize=20)
plt.plot(movements[1]); #1 for Apple

#we can see the inconsistency
```

```python
# we are goint to do for all of companies
from sklearn.preprocessing import Normalizer
normalizer = Normalizer()
norm_movements = normalizer.fit_transform(movements)
```

```python
# all values are in (-1,1)
print(norm_movements.min())
print(norm_movements.mean())
print(norm_movements.max())
```

```python
plt.figure(figsize=(20,8))

plt.subplot(1,2,1)

plt.title('Company: Amazon', fontsize= 20)
plt.xticks(fontsize=10)
plt.yticks(fontsize=20)
plt.xlabel('Date', fontsize=20)
plt.ylabel('Movement',fontsize=20)
plt.plot(norm_movements[0]) #0 for Amazon

plt.subplot(1,2,2)

plt.title('Company: Apple', fontsize= 20)
plt.xticks(fontsize=10)
plt.yticks(fontsize=20)
plt.xlabel('Date', fontsize=20)
plt.ylabel('Movement',fontsize=20)
plt.plot(norm_movements[1]); #1 for Apple
# After normalized
```

### Creating Pipline

```python
from sklearn.pipeline import make_pipeline
from sklearn.preprocessing import Normalizer
from sklearn.cluster import KMeans

normalizer = Normalizer()

#Create K-means model

kmeans = KMeans(n_clusters=5, max_iter=1000, random_state=1)

#make pipeline for combaning our normalizer and kmeans model

pipeline = make_pipeline(normalizer,kmeans)

#fit pipeline to daily stock movements 
pipeline.fit(movements)
predictions = pipeline.predict(movements) # 0 to 4
```

- The Companies and their respective clusters are displayed below

```python
df1 = pd.DataFrame({'Cluster':predictions, 'Companies':list(companies_list)}).sort_values(by = ['Cluster'], axis =0)
```

```python
#reducing deminsionalty
from sklearn.decomposition import PCA

normalizer = Normalizer()

pca_data = PCA(n_components=2)

kmeans = KMeans(n_clusters=5, max_iter=1000)

pipeline = make_pipeline(normalizer, pca_data, kmeans)

pipeline.fit(movements)
prediction = pipeline.predict(movements)

df2 = pd.DataFrame({'Cluster':predictions, 'Companies':list(companies_list.keys())}).sort_values(by = ['Cluster'], axis = 0)
df2
```

```python
from sklearn.decomposition import PCA

pca_data = PCA(n_components=2).fit_transform(norm_movements)
#define step size of mesh
h=0.05

#plot the desicion boundary
x_min, x_max = pca_data[:,0].min()-1, pca_data[:,0].max()+1
y_min, y_max = pca_data[:,1].min()-1, pca_data[:,1].max()+1
xx, yy = np.meshgrid(np.arange(x_min,x_max,h),np.arange(y_min,y_max,h))

#obtain labels for each point in the mesh using our trained model
kpredictions = kmeans.predict(np.c_[xx.ravel(),yy.ravel()])

#Put the results into a color plot
kpredictions = kpredictions.reshape(xx.shape)

#define color plot
cmap = plt.cm.Paired


# plotting figure
plt.clf()
plt.figure(figsize=(10,10))
plt.imshow(kpredictions, interpolation = 'nearest', extent=(xx.min(),xx.max(),yy.min(),yy.max()),cmap = cmap, aspect = 'auto',origin='lower')
plt.plot(pca_data[:,0], pca_data[:,1], 'k.', markersize= 5)

#Plot centroid of each cluster as a white x
centroids = kmeans.cluster_centers_
plt.scatter(centroids[:,0], centroids[:,1],marker = 'x', s=169, linewidths = 3, color='w', zorder =10)
plt.title('K-Means Clustering on stock market movements [PCA-Reduced data]')
plt.xlim(x_min,x_max)
plt.ylim(y_min,y_max)
plt.show();
```
